#### function to the venues from all neighborhoods in Bangkok

```python
def getNearbyVenues(names, latitudes, longitudes, radius):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&query={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            query,
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()['response']['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues )
```

#### 2.4 How to find the Potential Location
After request, get and merge all dataset together, we will get table of number of Pet Store and number of Household by each Neighborhood. To find the potential in each Neighborhood as follows.
     
 (1)...Household Proportion = (number of Household)/(Total Household)
 
 (2)...Competitive Proportion = (number of Pet Store)/(Total Pet Store)
 
 (3)...Potential Index = normalized(House Proportion)*(1-normalized(Competitive Proportion))

We will compare Potential index and select Neighborhood which the highest index to recommend for setting up Pet Store.

#### 3.3 Data Analysis
3.3.1 Make New data frame

```python
newDF = BKKDF[['Neighborhood','Household']]
newDF.head(10)
```

3.3.2 Put Pet Store to each neighborhood

```python
newDF_1 = pd.merge(newDF, newGBKKDF, how = 'left', on = 'Neighborhood')
newDF_1.rename(columns ={'Venue': 'Pet Store'}, inplace = True)
newDF_1.head(10)
```

Above table show nunber of household and number of pet store in neighborhood 

3.3.3 Find Potential Location

```python
# replace 0 for Pet Store = NaN
newDF_1['Pet Store'].fillna(0, inplace=True)
# Find Household proportion
newDF_1['Household Proportion'] = newDF_1['Household']/newDF_1['Household'].sum()
# Find Competitive Ratio
newDF_1['Competitive Proportion'] = newDF_1['Pet Store']/newDF_1['Pet Store'].sum()
newDF_1.head(10)
```

3.3.4 Normalize Household Porportion and Competitive Proportion)

```python
from sklearn.preprocessing import minmax_scale
def normalized(data):
    normalized_data = minmax_scale(data)
    return normalized_data

newDF_1['Normalized(Household Proportion)'] = normalized(np.array(newDF_1['Household Proportion']))
newDF_1['Normalized(Competitive Proportion)'] = normalized(np.array(newDF_1['Competitive Proportion']))
# Find (1-Normalized(Competitive Proportion))
newDF_1['(1-Normalized(Competitive Proportion))'] = 1-newDF_1['Normalized(Competitive Proportion)']
newDF_1.head(10)
```

3.3.5 Potential Index

Potential Index = normalized(HouseHold Proportion)*(1-normalized(Competitive Proportion))

High Index is more potential than Low Index 

```python
newDF_1['Potential Index'] = newDF_1['Normalized(Household Proportion)']*newDF_1['(1-Normalized(Competitive Proportion))']       
newDF_1.sort_values('Potential Index', ascending = False, inplace = True)
newDF_1.reset_index(drop = True, inplace = True)
newDF_1.head(10)
```
